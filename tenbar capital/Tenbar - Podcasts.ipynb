```python
#import libraries
from bs4 import BeautifulSoup
import pandas as pd
import requests
import urllib
from urllib.request import Request, urlopen
from urllib.error import HTTPError
from urllib.parse import urlparse
import re
```

```python
Pod = pd.DataFrame(columns = ['Podcast Name','URL','Type'])
hdr = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.2; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.100 Safari/537.36'}
```

```python
#search keyword for podcast focus on venture capital
query1 = '"podcast" in venture capital'
google_url = "https://www.google.com/search?q=" + query1


response = requests.get(google_url ,headers = hdr)
soup = BeautifulSoup(response.text, "html.parser")
```

```python
for line in soup.find_all(class_="llgymd"):
    link = line.get('href')
    updated_link = "https://www.google.com"+ link
    title = link.split("=")[1].split("&")[0].replace("+", " ")
    response1 = requests.get(updated_link ,headers = hdr)
    soup1 = BeautifulSoup(response1.text, "html.parser")
    result_div = soup1.find_all('div',attrs = {'class': 'r'})
    link = result_div[0].find('a', href = True)['href']
    print(link)
    Pod = Pod.append({'Podcast Name': title,
                                'URL': link,
                                'Type': "Venture Capital"
                                }, ignore_index=True)
```
