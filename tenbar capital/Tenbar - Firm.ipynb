# Venture Capital  and Private Equity firms to fund impact investments
    

```python
#import libraries
from bs4 import BeautifulSoup
import pandas as pd
import requests
import urllib
from urllib.request import Request, urlopen
from urllib.error import HTTPError
from urllib.parse import urlparse
import re
```

```python
hdr = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.2; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.100 Safari/537.36'}
```

```python
#search keyword for social impact investment firm
query1 = 'Venture Capital OR Private Equity in Social Impact'
```

```python
# create a dataframe to restore the Bachelor data

Social_firm = pd.DataFrame(columns = ['Firm Information','Organization Name','URL','Type'])

google_url4 = "https://www.google.com/search?q=" + query1 + "&num=200"

response = requests.get(google_url4 ,headers = hdr)
soup = BeautifulSoup(response.text, "html.parser")

result_div = soup.find_all('div',attrs = {'class': 'r'})

for r in result_div:
    try:
        link = r.find('a', href = True)
        title = r.find('h3', attrs={'class':'LC20lb DKV0Md'}).get_text()
        print(link['href'])
        print (title)
        if link!= ''and title!= '': 
            text = urlparse(link['href']).netloc.split('.')
            Social_firm = Social_firm.append({'Firm Information': title,
                                'Organization Name':text[len(text)-2],
                                'Type': "Social Impact",
                                'URL': link['href']}, ignore_index=True)
    # Next loop if one element is not present
    except:
        continue

#search for 100+ results
google_url4 = google_url4+"&start=100"

response = requests.get(google_url4 ,headers = hdr)
soup = BeautifulSoup(response.text, "html.parser")

result_div = soup.find_all('div',attrs = {'class': 'r'})

for r in result_div:
    try:
        link = r.find('a', href = True)
        title = r.find('h3', attrs={'class':'LC20lb DKV0Md'}).get_text()
        if link!= ''and title!= '': 
            text = urlparse(link['href']).netloc.split('.')
            Social_firm = Social_firm.append({'Firm Information': title,
                                'Organization Name':text[len(text)-2],
                                'Type': "Social Impact",
                                'URL': link['href']}, ignore_index=True)
    # Next loop if one element is not present
    except:
        continue
                               

#search for 200+ results
google_url4 = google_url4+"&start=200"

response = requests.get(google_url4 ,headers = hdr)
soup = BeautifulSoup(response.text, "html.parser")

result_div = soup.find_all('div',attrs = {'class': 'r'})

for r in result_div:
    try:
        link = r.find('a', href = True)
        title = r.find('h3', attrs={'class':'LC20lb DKV0Md'}).get_text()
        if link!= ''and title!= '': 
            text = urlparse(link['href']).netloc.split('.')
            Social_firm = Social_firm.append({'Firm Information': title,
                                'Organization Name':text[len(text)-2],
                                'Type': "Social Impact",
                                'URL': link['href']}, ignore_index=True)
    # Next loop if one element is not present
    except:
        continue
```
