This Project is to research on all United States based Undergraduate/Graduate Degree granting Programs in Sustainability, Social Impact, ESG

Input Data - type of program, Program Specialization keyword , number of result 
Output Data - List of US based college with type of program and degree you picked

Logic: Create separate dataframe for each program and merge them together

```python
#import libraries
from bs4 import BeautifulSoup
import pandas as pd
import requests
import urllib
from urllib.request import Request, urlopen
from urllib.error import HTTPError
from urllib.parse import urlparse
```

```python
# Get first search keyword
keyword = pd.read_excel("Tenbar.xlsx", sheet_name="Keyword")
keyword 
```

```python
hdr = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.2; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.100 Safari/537.36'}
```

```python
# set up MBA search keyword and number of results expected

query1 = "United States "+keyword.iloc[0]['Program'] + " " + keyword.iloc[0]['Keyword'] + " site:edu -inurl:news -inurl:blog -inurl:blogs -inurl:story "
result_num_1 = keyword.iloc[0]['Number of results']
```

```python
# create a dataframe to restore the MBA data
MBA_data = pd.DataFrame(columns = ['University-short','MBA Program','MBA Website'])

google_url1 = "https://www.google.com/search?q=" + query1 + "&num=" + str(result_num_1)

response = requests.get(google_url1 ,headers = hdr)
soup = BeautifulSoup(response.text, "html.parser")

result_div = soup.find_all('div',attrs = {'class': 'r'})

for r in result_div:
    try:
        link = r.find('a', href = True)
        title = r.find('h3', attrs={'class':'LC20lb DKV0Md'}).get_text()
        if link!= ''and title!= '': 
            text = urlparse(link['href']).netloc.split('.')
            MBA_data = MBA_data.append({'University-short':text[len(text) - 2],
                        'MBA Program': title,
                        'MBA Website': link['href']}, ignore_index=True)
    # Next loop if one element is not present
    except:
        continue


#search for 100+ results
google_url1 = google_url1+"&start=100"
response = requests.get(google_url1 ,headers = hdr)
soup = BeautifulSoup(response.text, "html.parser")

result_div = soup.find_all('div',attrs = {'class': 'r'})

for r in result_div:
    try:
        link = r.find('a', href = True)
        title = r.find('h3', attrs={'class':'LC20lb DKV0Md'}).get_text()
        if link!= ''and title!= '': 
            text = urlparse(link['href']).netloc.split('.')
            MBA_data = MBA_data.append({'University-short':text[len(text) - 2],
                        'MBA Program': title,
                        'MBA Website': link['href']}, ignore_index=True)
    # Next loop if one element is not present
    except:
        continue
```

```python
Merge_data = MBA_data.merge(Cert_data, on='University-short', how='outer',sort=True)
Merge_data= Merge_data.merge(ugrad_data, on='University-short', how='outer',sort=True)
Merge_data
```

```python
# Add college informations
college = pd.read_excel("Tenbar.xlsx", sheet_name="US College")
df1 = Merge_data.merge(college[['College','City','State','University-short']], on='University-short', how='left')
```

```python
# Check on final information
df2=df1[['College','City','State','MBA Program','MBA Website','Certificate Program','Certificate Website',
        'Bachelor Program','Bachelor Website']]
df2
```

```python
bussiness = pd.read_excel("Tenbar.xlsx", sheet_name="Business School")
```

```python
# add business school information
final_df = df2.merge(bussiness[['College','Business School Name']], on='College', how='left')
final_df=final_df[['College','City','State','Business School Name','MBA Program','MBA Website','Certificate Program','Certificate Website','Bachelor Program','Bachelor Website']]
df = final_df[pd.notnull(final_df['College'])]
df
```

```python
#write to excel
with pd.ExcelWriter('Tenbar.xlsx',engine="openpyxl", mode='a') as writer:
    df.to_excel(writer, sheet_name='Final Data')
```
