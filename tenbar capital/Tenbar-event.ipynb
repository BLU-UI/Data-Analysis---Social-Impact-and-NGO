# ESG/Sustainability/Impact Investing Related event search

```python
#import libraries
from bs4 import BeautifulSoup
import pandas as pd
import requests
import urllib
from urllib.request import Request, urlopen
from urllib.error import HTTPError
from urllib.parse import urlparse
import re
```

```python
hdr = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.2; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.100 Safari/537.36'}
```

```python
#search keyword
query = 'Conference OR Webinar OR Summit in ESG OR Sustainability OR Impact Investment "2020" OR "2021"'
```

```python
# create a dataframe to restore the ESG/Sustainable/social impact events

df = pd.DataFrame(columns = ['Conference/Summit/Webinar Information','URL','Organization Type','Organization Name'])

google_url = "https://www.google.com/search?q=" + query  + "&num=250"

response = requests.get(google_url ,headers = hdr)
soup = BeautifulSoup(response.text, "html.parser")

result_div = soup.find_all('div',attrs = {'class': 'r'})

for r in result_div:
    try:
        link = r.find('a', href = True)
        title = r.find('h3', attrs={'class':'LC20lb DKV0Md'}).get_text()
        print(link['href'])
        print (title)
        if link!= ''and title!= '': 
            text = urlparse(link['href']).netloc.split('.')
            if (text[len(text)-1] == 'edu'):
                o_type = "Educational Institution"
            elif (text[len(text)-1] == 'net'):
                o_type = "Networking Technologies"
            elif (text[len(text)-1] == 'org'):
                o_type = "Organization"
            elif (text[len(text)-1] == 'com'):
                o_type = "Company"
            elif (text[len(text)-1] == 'io'):
                o_type = "Startup"
            else:
                o_type = "Not Known"
            df = df.append({'Conference/Summit/Webinar Information': title, 'Organization Name':text[len(text)-2],
                            'URL': link['href'], 'Organization Type': o_type}, ignore_index=True)
    # Next loop if one element is not present
    except:
        continue

#search for 100+ results
google_url = google_url+"&start=100"

response = requests.get(google_url ,headers = hdr)
soup = BeautifulSoup(response.text, "html.parser")

result_div = soup.find_all('div',attrs = {'class': 'r'})

for r in result_div:
    try:
        link = r.find('a', href = True)
        title = r.find('h3', attrs={'class':'LC20lb DKV0Md'}).get_text()
        print(link['href'])
        print (title)
        if link!= ''and title!= '': 
            text = urlparse(link['href']).netloc.split('.')
            if (text[len(text)-1] == 'edu'):
                o_type = "Educational Institution"
            elif (text[len(text)-1] == 'net'):
                o_type = "Networking Technologies"
            elif (text[len(text)-1] == 'org'):
                o_type = "Organization"
            elif (text[len(text)-1] == 'com'):
                o_type = "Company"
            elif (text[len(text)-1] == 'io'):
                o_type = "Startup"
            else:
                o_type = "Not Known"
            df = df.append({'Conference/Summit/Webinar Information': title, 'Organization Name': text[len(text)-2],
                            'URL': link['href'], 'Organization Type': o_type}, ignore_index=True)
    # Next loop if one element is not present
    except:
        continue
```
