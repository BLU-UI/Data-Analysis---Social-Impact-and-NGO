```python
#import libraries
import urllib
from urllib.request import Request, urlopen
from urllib.error import HTTPError
from urllib.parse import urlparse
from googlesearch import search
from bs4 import BeautifulSoup
import pandas as pd
import ssl
ssl._create_default_https_context = ssl._create_unverified_context
import re
import requests
```

```python
# Get first search keyword
keyword = pd.read_excel("Tenbar.xlsx", sheet_name="Keyword")
keyword 
```

```python
hdr = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.2; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.100 Safari/537.36'}
```

```python
# set up MBA search keyword and number of results expected

query1 = "United States "+keyword.iloc[0]['Program'] + " " + keyword.iloc[0]['Keyword'] + " site:edu -inurl:news -inurl:blog -inurl:blogs -inurl:story "
result_num_1 = keyword.iloc[0]['Number of results']
google_url1
```

```python
# create a dataframe to restore the MBA data
MBA_data = pd.DataFrame(columns = ['University-short','MBA Program','MBA Website'])

google_url1 = "https://www.google.com/search?q=" + query1 + "&num=" + str(result_num_1)

response = requests.get(google_url1 ,headers = hdr)
soup = BeautifulSoup(response.text, "html.parser")

result_div = soup.find_all('div',attrs = {'class': 'r'})

for r in result_div:
    try:
        link = r.find('a', href = True)
        title = r.find('h3', attrs={'class':'LC20lb DKV0Md'}).get_text()
        if link!= ''and title!= '': 
            text = urlparse(link['href']).netloc.split('.')
            MBA_data = MBA_data.append({'University-short':text[len(text) - 2],
                        'MBA Program': title,
                        'MBA Website': link['href']}, ignore_index=True)
    # Next loop if one element is not present
    except:
        continue


#search for 100+ results
google_url1 = google_url1+"&start=100"
response = requests.get(google_url1 ,headers = hdr)
soup = BeautifulSoup(response.text, "html.parser")

result_div = soup.find_all('div',attrs = {'class': 'r'})

for r in result_div:
    try:
        link = r.find('a', href = True)
        title = r.find('h3', attrs={'class':'LC20lb DKV0Md'}).get_text()
        if link!= ''and title!= '': 
            text = urlparse(link['href']).netloc.split('.')
            MBA_data = MBA_data.append({'University-short':text[len(text) - 2],
                        'MBA Program': title,
                        'MBA Website': link['href']}, ignore_index=True)
    # Next loop if one element is not present
    except:
        continue
```
